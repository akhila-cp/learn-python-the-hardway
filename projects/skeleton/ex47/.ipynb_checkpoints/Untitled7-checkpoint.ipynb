```python
#Rahna KM #rahnarazak@gmail.com
#Data set consist of two inputs and a output.Hence we have to apply multiple linear regression.
#Multiple linear model steps are:
#1. Data preprocessing
#   -Importing libraries and data set.
#   -Finding correlation
#   -avoid dummy variable trap
#   -Splitting dataset into train and test set.
#2. Fitting multiple linear regression to training data.
#3. Compute root-mean-square-error and display results

import numpy
from numpy import arange
from matplotlib import pyplot
import pandas as pd
from pandas import read_csv
from pandas import set_option
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

#Reading dataset
filename = ("/Training_Data.csv")
print('Parsing dataset from CSV....')
names = ['X1','X2','Y']
dataset = read_csv(filename, names=names)
print(f"dataset shape:{dataset.shape}")
print(f"dataset dtypes:\n{dataset.dtypes}")
dataset = dataset.iloc[1:]
dataset = dataset.astype(float)
dataset.isnull().sum().max()

#Plotting of figure
fig = plt.figure()
res = stats.probplot(dataset['X1'], plot=plt)
res = stats.probplot(dataset['X2'], plot=plt)
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection="3d")
z_points = dataset['Y']
x_points = dataset['X1']
y_points = dataset['X2']
ax.scatter3D(x_points, y_points, z_points, c=z_points, cmap='hsv');
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
plt.show()

#Training and testing data
X = dataset.values[:,0:2]
Y = dataset.values[:,2]
validation_size = 0.20
seed = 7
print('Creating Test and Train data from dataset....')
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=validation_size, random_state=seed)
sc = StandardScaler()
train_x = sc.fit_transform(X_train)
test_x = sc.transform(X_validation)
lr = LinearRegression()
lr.fit(X_train,Y_train)
predict_test = lr.predict(X_validation)
predict_train = lr.predict(X_train)

#Results
print('Results')
print('Intercept: \n', lr.intercept_)
print('Coefficients: \n', lr.coef_)
print('RMSE for test Data:',numpy.sqrt(mean_squared_error(Y_validation,predict_test)))
print('RMSE for train Data:',numpy.sqrt(mean_squared_error(Y_train,predict_train)))

```
